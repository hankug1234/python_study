```python
import cv2
import numpy as np
import sys
```

```python
img = cv2.imread("./fig/puppy.bmp",cv2.IMREAD_COLOR)

if img is None:
    print("image read failed")
    sys.exit()

cv2.namedWindow("img",cv2.WINDOW_AUTOSIZE)
cv2.imshow("img",img)
while True:
    key = cv2.waitKey()
    if key == 27 or key == ord("q"):
        break
        
cv2.destroyAllWindows()
```

```python
import os 
import cv2
import numpy as np
import sys 
```

```python
os.listdir("./fig/images")
```

```python
img_lists = os.listdir("./fig/images")
img_files = ["./fig/images/"+ img for img in img_lists]
print(img_files)
```

```python
cv2.namedWindow("scenery",cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scenery',cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN)
for i in img_files:
    img = cv2.imread(i)
    
    if img is None:
        print("img read failed")
        break
    
    cv2.imshow('scenery',img)
    
    if cv2.waitKey(3000) == 27:
        break
cv2.destroyAllWindows()        
```

```python
import glob
import os 
import cv2
import numpy as np
import sys 
```

```python
astra_list = glob.glob("./fig/test/*.*")
cv2.namedWindow("scenery",cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scenery',cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN)
state = 0
while True:
    for i in astra_list:
        img = cv2.imread(i)
    
        if img is None:
            print("img read failed")
            break
    
        cv2.imshow('scenery',img)
        
        if cv2.waitKey(1000) == 27:
            state = 1
            break
    if state:
        break
cv2.destroyAllWindows()     
```
