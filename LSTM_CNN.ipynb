```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Embedding, LSTM, Conv1D, MaxPooling1D
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.callbacks import EarlyStopping
import attention as Attention    #With Attention

import numpy as np
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
x_train = sequence.pad_sequences(x_train, maxlen=500)
x_test = sequence.pad_sequences(x_test, maxlen=500)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
modelA = Sequential()
modelA.add(Embedding(5000,100))
modelA.add(Dropout(0,5))
modelA.add(Conv1D(64,5,padding='valid',activation='relu',strides=1))
modelA.add(MaxPooling1D(pool_size=4))
modelA.add(LSTM(64, return_sequences=True))
modelA.add(Dense(1))
modelA.add(Activation("sigmoid"))

modelA.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=3)

history = modelA.fit(x_train,y_train, batch_size=40, epochs=100, validation_split=0.25, callbacks=[early_stopping_callback])
```
