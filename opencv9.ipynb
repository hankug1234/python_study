```python
import cv2
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```python
src = cv2.imread("./fig/puppy.bmp",0)

hist = cv2.calcHist([src],[0],None,[256],[0,256])

plt.plot(hist)

plt.show()
```

```python
src = cv2.imread("./fig/puppy.bmp",1)

hist_b = cv2.calcHist([src],[0],None,[256],[0,256])
hist_g = cv2.calcHist([src],[1],None,[256],[0,256])
hist_r = cv2.calcHist([src],[2],None,[128],[0,256])

plt.plot(hist_b)
plt.plot(hist_g)
plt.plot(hist_r)

plt.show()
```

```python
src = cv2.imread('./fig/puppy.bmp',1)
src_hsv = cv2.cvtColor(src,cv2.COLOR_RGB2HSV)

b,g,r = cv2.split(src)
h,s,v = cv2.split(src)

cv2.imshow("b",b)
cv2.imshow("g",g)
cv2.imshow("r",r)

cv2.imshow("h",h)
cv2.imshow("s",s)
cv2.imshow("v",v)

cv2.waitKey()
cv2.destroyAllWindows()
```
